#**Relations based on Sub-Catagory**


1. Profit with Discount and Quantity

```python
fig,axes = plt.subplots(1,2,figsize=(20,5))
data.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Profit & Discount')
data.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Profit & Quantity')
plt.show()
```

2. Relation of Profit and sales

```python
fig,axes = plt.subplots(1,2,figsize=(20,5))

data.groupby('Sub-Category')['Profit', 'Sales'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Profit and Sales')

plt.title('Profit VS Sales')
sns.barplot(data=data.groupby('Sub-Category')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
                                                               
plt.xticks(rotation=90)
plt.show()
```

3.  Relation of Sales and Quantity

```python
fig,axes = plt.subplots(1,2,figsize=(20,5))

data.groupby('Sub-Category')['Sales', 'Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Sales and Quantity')

plt.title('Quantity VS Sales')
sns.barplot(data=data.groupby('Sub-Category')['Sales','Quantity'].agg(sum),x='Sales',y='Quantity',ax=axes[1])
                                                               
plt.xticks(rotation=90)
plt.show()
```

# **Profit and Sales Based on various factors**

# Region-wise Profit and Sales

```python
df_regP = data.groupby(['Region'])[['Profit']].sum()
df_regS = data.groupby(['Region'])[['Sales']].sum()

fig,ax = plt.subplots(1,2,figsize=(12,5))

df_regP.plot(kind='bar',ax=ax[0]).set_title('Region profit')
df_regS.plot(kind='bar',ax=ax[1]).set_title('Region Sales')
```

# State-wise Profit and Sales

```python
df_stateP = data.groupby('State')[['Profit']].sum().sort_values('Profit')
df_stateS = data.groupby('State')[['Sales']].sum().sort_values('Sales')

fig,ax = plt.subplots(1,2,figsize=(25,15))

df_stateP.plot(kind='barh',ax=ax[0], ylabel= 'Profit', xlabel= 'State').set_title('State Profit')
df_stateS.plot(kind='barh',ax=ax[1], ylabel= 'Sales', xlabel= 'State').set_title('State Sales')
```

Here we have visualised profit faced by various states.

* The `highest profit` earning states are

1. CALIFORNIA
2. NEW YORK
3. WASHINGTON

There are negative profit values in some states, so we can say that they have suffered loss.

* These states have suffered `major loss`

1. TEXAS
2. OHIO
3. PENNSYLVANIA
4. ILLINOIS

*  States with `highest Sales`
1.  CALIFORNIA
2.  NEW YORK 
3.  TEXAS


*  State with `lowest Sales`
1. NORTH DAKOTE
2. WEST VIRGINIA
3. MAINE

# City-wise Profit and Sales

```python
p1 = data.groupby(['City'])[['Profit']].sum().sort_values('Profit').head(10)
p2 = data.groupby(['City'])[['Profit']].sum().sort_values('Profit').tail(10)

fig,ax = plt.subplots(1,2,figsize=(15,5))

p1.plot(kind='bar',ax=ax[0]).set_title('Cities in loss')
p2.plot(kind='bar',ax=ax[1]).set_title('Cities in profit')
```

```python
p3 = data.groupby(['City'])[['Sales']].sum().sort_values('Sales').head(10)
p4 = data.groupby(['City'])[['Sales']].sum().sort_values('Sales').tail(10)

fig,ax = plt.subplots(1,2,figsize=(15,5))

p3.plot(kind='bar',ax=ax[0]).set_title('Cities in loss')
p4.plot(kind='bar',ax=ax[1]).set_title('Cities in profit')
```

Here we have visualised profit and sales based on Cities
*  Cities with `highest profit`
1. New York City
2. Los Angeles
3. Seattle

* Cities with `highest loss`
1. Philadelphia
2. Houston
3. San Antonio

* Cities with `highest Sales`
1. New York City
2. Los Angeles
3. Seattle


* Cities with `Lowest Sales`
1. Abilene
2. Elyria
3. Jupiter


# Catagory-wise Profit and Sales

```python
df_C1 = data.groupby(['Category'])[['Profit']].sum().sort_values('Profit')
df_C2 = data.groupby(['Category'])[['Sales']].sum().sort_values('Sales')

fig,ax = plt.subplots(1,2,figsize=(15,5))

df_C1.plot(kind='bar',ax=ax[0]).set_title('Catagory wise Profit')
df_C2.plot(kind='bar',ax=ax[1]).set_title('Catagory wise Sales')
```


*  The Technology catagory has the highest profit as well as highest Sales

# Sub-Catagory wise Profit and Sales

```python
df_SC1 = data.groupby(['Sub-Category'])[['Profit']].sum().sort_values('Profit')
df_SC2 = data.groupby(['Sub-Category'])[['Sales']].sum().sort_values('Sales')

fig,ax = plt.subplots(1,2,figsize=(15,5))

df_SC1.plot(kind='bar',ax=ax[0]).set_title('Sub-catagory wise Profit')
df_SC2.plot(kind='bar',ax=ax[1]).set_title('Sub-catagory wise Sales')
```

Here we can see that,

* Tables, Bookcases and Supplies, these sub-catagories have faced losses

* Although the sales of Tables are pretty high it has suffered the maximum loss

* The Copiers sub-catagory has the highest profits

* The Phones sub-catagory has the highest sales
